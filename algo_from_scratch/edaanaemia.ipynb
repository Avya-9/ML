```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('anaemia.csv')
df
```

```python
df.shape
```

```python
print(df['Sex'].unique())
```

```python
df['Sex'] = df['Sex'].str.strip()
```

```python
dummies=pd.get_dummies(df['Sex'],drop_first=True)
dummies
```

```python
dummies["M"] = dummies["M"].astype(int)
dummies.head()
```

```python
df=pd.concat([df,dummies],axis='columns')
df.head()
```

```python
df.drop(['Sex'],axis='columns',inplace=True)
```

```python
df.head()
```

```python
df.to_csv('clean_anaemia.csv', index=False)
```
